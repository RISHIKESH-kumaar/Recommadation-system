```python
import ast
```

```python
# Convert string representation of list of dictionaries into a list of 'name' values
def convert(text):
  l=[]
  for i in ast.literal_eval(text):
      l.append(i['name'])
  return l
```

```python

movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
#extracting the name of each director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(4)
```

```python
#removing spaceses from each string in the list
def collapse(l):
  l1=[]
  for i in l:
    l1.append(i.replace(' ',''))
  return l1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
#splitting each sentence into invidual words

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#now adding overview,generes,keywords,cast,crew
movies['tag']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
#converting list into single string
new['tag'] = new['tag'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

vectorization (converting text into numbers)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tag']).toarray()
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
#it finds closest vector

similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python

def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(enumerate(similarity[index]), key=lambda x: x[1], reverse=True)
    
    recommendations = []
    for i in distances[1:6]:
        recommendations.append(new.iloc[i[0]].title)
    
    return recommendations
```
